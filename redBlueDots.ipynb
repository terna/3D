```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import numpy.random as npr

fig, ax = plt.subplots()
xdata1, ydata1, xdata2, ydata2 = [], [], [], []
ln = plt.plot([],[],'ro',[],[],'bo')

def init():
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    return ln

def update(frame):
    xdata1=list(npr.random(10))
    ydata1=list(npr.random(10))
    xdata2=list(npr.random(8))
    ydata2=list(npr.random(8))
    #print(frame)
    
    ln[0].set_data(xdata1, ydata1)
    ln[1].set_data(xdata2, ydata2)
    return ln

# pars https://matplotlib.org/api/_as_gen/matplotlib.animation.FuncAnimation.html
# interval - Delay between frames in milliseconds. Defaults to 200.
ani = FuncAnimation(fig, update, frames=list(range(100)),
                    init_func=init, blit=True, repeat=False, interval=50) # blit=False
                    
#ani.save("boh.mp4") 
plt.show()
```
